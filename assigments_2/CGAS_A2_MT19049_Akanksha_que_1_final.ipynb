step 1:

Assign uniform value between 0 to 1 to all the ingredient in nature basket(I).

1 to 4 steps of algorithm

```
import random
fitness_ingred={}
for i in I:
  while True:
    x=random.uniform(0,1)
    if x in fitness_ingred.values():
      continue
    else:
      fitness_ingred[i]=x
      break
```

7.R_0=randomly sample S_bar ingredients n times from I_0.

```
R_0=[]
S_bar=10
n=500
for i in range(0,n):
    recip=[]
    while len(recip)!=S_bar:
      index=random.randint(0,len(I_0)-1)
      if I_0[index] in recip:
         continue
      else:
         recip.append(I_0[index])
    R_0.append(set(recip)) 
```

Recipie size distribution code:

```
def recipie_size_plot(recipes,epoch):  
  print("Epoch: ",epoch)
  rsize=[]
  for i in recipes:
      rsize.append(len(i))

  y_axis=[]
  for i in range(0,max(rsize)+1):
      y_axis.append(rsize.count(i)*100/len(rsize))  
  recp_y.append(y_axis)
  plt.plot(y_axis, label='recipe(ingredient generated) size distribution')
  plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
  plt.xlabel(' recipe size')
  plt.ylabel('Percentage( normalised frequency)')
  plt.title('Recipe size distribution')
  plt.show()
```

Frequncy rank distribution code:

```
import math
def frquency_rank(R_0,I_0,epoch):

    ingred={k:0 for k in I_0}
    for re in R_0:
        for ingre in re:
                ingred[ingre]+=1
    # figure, axes = plt.subplots(figsize = (8,6))
    ingred={k: v for k, v in sorted(ingred.items(),key=lambda item: item[1],reverse=True)}

    b=list(ingred.values())
    d=0
    for i in b:
        d+=i
    a=[ x for x in range(1,len(b)+1)]
    b1=[y/d for y in b]
    A_graph.append(a)
    B_graph.append(b1)
```
